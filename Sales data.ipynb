```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('Sales Data.csv')
data = df.copy()
data.head(20)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.describe()
```

```python
data = data.drop(['Unnamed: 0'],axis =1)
data.head()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
data['Month'] = data['Order Date'].dt.month
data['Hour'] = data['Order Date'].dt.hour

data['Day of Week'] = data['Order Date'].dt.day_name()
data.head()
```

```python
data.info()
```

```python
product = data['Product'].value_counts().head(10)
product
```

```python
city = data['City'].value_counts()
city
```

```python
sns.barplot(x = 'Day of Week',y = 'Sales',data=data,estimator = sum, palette = 'rocket')
plt.title('Sales of day of week')
plt.xlabel('Day of week')
plt.ylabel('Sales(All values are in Millions)')
```

```python
col = data.select_dtypes(include = ['number'])
corr = col.corr()
sns.heatmap(corr,annot = True,cmap = 'coolwarm')
```

```python
sns.barplot(x = 'City', y = 'Sales',data = data, estimator = sum, palette = 'rocket')
plt.xticks(rotation = 45)
plt.ylabel('Sales(All values are in Millions)')
plt.title('Sales by cities')
plt.xlabel('City')
#most product puchased by the people of san francisco
```

```python
sns.barplot(x='Month', y='Sales', data=data, estimator=sum,palette = 'rocket')
plt.title('Total Sales by Month')
plt.xlabel('month')
plt.ylabel('Sales(All values are in Millions)')
# in the month of december sales is more
```

```python
sns.barplot(x = 'Sales' ,y= 'Product' ,data = data,estimator = sum,palette = 'crest')
plt.title('Sales of product')
plt.xlabel('Sales(All values are in Millions)')
plt.ylabel('Product')
```

```python
sns.barplot(x = 'Quantity Ordered', y ='Product',data = data,estimator = sum,palette = 'crest')
plt.title('Quantity ordered of each product')
plt.xlabel('Quantity Ordered')
plt.ylabel('Product')
```

```python
sns.barplot(x = 'Price Each' , y = 'Product',data = data, palette = 'crest')
plt.title('Price of each product')
plt.xlabel('Price each')
plt.ylabel('Product')
```

```python
monthly_sales = data.groupby('Month')['Sales'].sum().reset_index()
monthly_sales
```

```python
plt.plot(monthly_sales['Month'] , monthly_sales['Sales'],marker = 'o')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Sales')
```
